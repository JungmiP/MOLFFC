```python
import os
import pathlib
import cv2
from tqdm.notebook import tqdm
import shutil
import math
```

```python
#동영상 1개당 20프레임씩 추출
MAX_FRAME_NUM = 20
```

```python
#동영상에서 추출한 프레임을 저장할 폴더
frame_path = f"./frame_data"

#pathlib.Path(frame_path): 프레임을 저장할 폴더의 정보를 알아낼 객체 리턴
frame_save_path = pathlib.Path(frame_path)

#frame_save_path.exists(): 프레임을 저장할 폴더 data/frame이 존재하면
# if frame_save_path.exists():
#     #프레임을 저장할 폴더 data/frame 삭제
#     shutil.rmtree(frame_path)
```

```python
#학습 동영상 파일들의 경로를 저장할 리스트
mp4_path_list = []

#동영상 파일의 경로
mp4_root_path = "./video_data"

#mp4_root_path 폴더에 포함된 파일들 리턴
for(path, dir, files) in os.walk(mp4_root_path):
    #폴더 경로
    print("path=", path)
    #서브 폴더명
    print("dir=", dir)
    #폴더에 저장된 파일 리스트
    print("files=", files)
    print("=" * 100)

    #path에 포함된 파일 리스트에서 파일명을 filename에 대입
    for filename in files:
        print("filename=", filename)
        #filename을 파일명과 확장자로 분리
        spname = os.path.splitext(filename)
        print("spname=", spname)
        #파일명의 확장자 리턴
        ext = spname[-1]
        print("ext=", ext)
        print("=" * 100)

        #확장자가 .mp4일 때
        if ext == '.mp4':
            #파일 경로 path 와 filename 결합
            file_path = path + "/"+filename
            #mp4_path_list에 확장자가 mp4인 파일의 경로 추가
            mp4_path_list.append(file_path)
        print("=" *100)
```

```python
#mp4 파일 경로 조회
mp4_path_list
```

```python
for mp4_path in mp4_path_list:
    print("mp4_path1=",mp4_path)
    
    #mp4_path에 \\를  /로 변경
    mp4_path = mp4_path.replace("\\","/")
    print("mp4_path2=",mp4_path)
    #mp4_path_list에 저장된 mp4 파일 경로를 / 로 분리
    sp1 = mp4_path.split("/")

    #폴더명 리턴
    class_name = sp1[-2]
    print("class_name=",class_name)

    frame_class_path = f"{frame_path}/{class_name}"
    print("frame_class_path=", frame_class_path)
    
    #pathlib.Path(frame_path): 프레임을 저장할 폴더의 정보를 알아낼 객체 리턴
    frame_save_path = pathlib.Path(frame_class_path)

    #frame_save_path.exists(): 프레임을 저장할 폴더 ./frame_data/class_name 이 존재하지 않으면
    if not frame_save_path.exists():
        #프레임 저장할 폴더 ./frame_data/class_name 생성
        os.makedirs(frame_class_path)
    
    #파일명 리턴
    video_name = sp1[-1]
    print("video_name=",video_name)
    
    #확장자 .mp4를 제외한 파일명 리턴
    video_name_not_ext = video_name[:-4]
    print("video_name_not_ext=",video_name_not_ext)
    
    #mp4 파일의 정보를 가져올 객체 생성
    capture = cv2.VideoCapture(mp4_path)

    #mp4 파일의 프레임 개수 조회
    frame_count = int(capture.get(cv2.CAP_PROP_FRAME_COUNT))
    print("frame_count=", frame_count)

    #동영상 1개당 20프레임씩 저장하기 위해
    #전체 동영상 개수 frame_count를 MAX_FRAME_NUM으로 나눔
    skip_num = frame_count / MAX_FRAME_NUM
    #나눈 값을 내림처리
    skip_num = math.floor(skip_num)
    print("skip_num=",skip_num)

    #동영상의 프레임 번호
    frame_num = 0
    #저장된 jpg 순서
    jpg_num = 0

    while(True):
        #동영상의 프레임을 읽어서 저장
        success, frame = capture.read()

        #success : 동영상읽기 성공
        #(frame_num (동영상 프레임 순서)  % skip_num == 0) : 동영상이 skip_num 번째 프레임
        #(jpg_num < MAX_FRAME_NUM) : 동영상 번호가 MAX_FRAME_NUM (20) 미만
        if success & (frame_num % skip_num == 0) & (jpg_num < MAX_FRAME_NUM):
            jpg_path = f'{frame_class_path}/{video_name_not_ext}_{jpg_num}.jpg'
            print("jpg_path=",jpg_path)
            cv2.imwrite(jpg_path, frame)
            jpg_num = jpg_num + 1

        if not success:
            break

        frame_num = frame_num + 1

    #비디오 파일 닫기
    capture.release()
    print("=" * 100)
```
