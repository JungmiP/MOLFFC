```python
IM_HEIGHT = 224
IM_WIDTH = 224
ORIGINAL_HEIGHT = 26
ORIGINAL_WIDTH = 34
```

```python
len(df)
```

```python
count = len(df)
```

```python
count
```

```python
imgs = np.zeros((count,IM_HEIGHT,IM_WIDTH),dtype=np.uint8)
```

```python
imgs.shape
```

```python
imgs
```

```python
tgs = np.zeros((count,1) ,dtype=np.float32)
```

```python
tgs.shape
```

```python
tgs
```

```python
for index in range(count):
    img = df.loc[index,"image"]
    img = img.replace("[","").replace("]","").replace(" ","")
    img = img.split(",")
    im = np.array(img,dtype=np.uint8)
    im = im.reshape((ORIGINAL_HEIGHT, ORIGINAL_WIDTH))

    im = cv2.resize(im, dsize=(IM_HEIGHT, IM_WIDTH), interpolation=cv2.INTER_LINEAR)

    imgs[index] = im

    tag = df.loc[index,"state"]

    if tag == 'open':
        tgs[index] = 1
    else:
        tgs[index] = 0
```
