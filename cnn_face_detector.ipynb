```python
import cv2
import dlib
import glob
import matplotlib.pyplot as plt
```

```python
categories = ["focus_on", "boring", "sleeping"]
```

```python
#모델 불러오기
cnn_face_detector = dlib.cnn_face_detection_model_v1("./mmod_human_face_detector.dat")

for category in categories:
    path = f"./frame_data/{category}/*.jpg"


    #이미지 파일 목록 불러오기
    file_list = glob.glob(path)
    #print("file_list : ", file_list)
    
    for index, file in enumerate(file_list):
        
        image = None
        print("file=", file)
        image = cv2.imread(file)
        print("image.shape=", image.shape)
        try:
            
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            file_name = file.split(sep="\\")[-1]
            face_detections = cnn_face_detector(image, 1)
            
            left, top, right, bottom = face_detections[0].rect.left(), face_detections[0].rect.top(), face_detections[0].rect.right(), face_detections[0].rect.bottom()
            
            left -= 50
            top -= 50
            right += 50
            bottom += 50
            
            cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)
            
            face_img = image[top:bottom, left:right]
            #print(face_img)
            
            #plt.imshow(face_img)
            
            save_file_path = f"./preprocessed_dataset/train/{category}/{file_name}"
            face_img = cv2.cvtColor(face_img, cv2.COLOR_RGB2BGR)
            cv2.imwrite(save_file_path, face_img)
            print("="*100)
        except Exception as e:    
            print('예외가 발생했습니다.', e)
            print("image=",image)
            print("="*100)
```
